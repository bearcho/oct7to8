```python
import csv
with open('bastket.csv','r',encoding='UTF-8-sig') as cf:
    transactions = []
    r = csv.reader(cf)
    for row in r:
        transactions.append(row)
```

```python
transactions
```

```python
from apyori import apriori
```

```python
rules = apriori(transactions,min_support=0.1,min_confidence=0.1)
rules
```

```python
results = list(rules)
```

```python
results[15]
```

```python
for row in results:
    support = row[1]
    ordered_statics = row[2]
    for ordered_item in ordered_statics:
        lhs = [x for x in ordered_item[0]]
        rhs = [x for x in ordered_item[1]]
        confidence = ordered_item[2]
        lift = ordered_item[3]
        print(f"{lhs} -> {rhs} \t {support} \t {confidence} \t {lift}")
    
```

```python
import requests
rss_url = "http://fs.jtbc.joins.com/RSS/economy.xml"
jtbc_economy = requests.get(rss_url)
```

```python
from bs4 import BeautifulSoup
economy_new_list = BeautifulSoup(jtbc_economy.content,'xml')
```

```python
link_list = economy_new_list.select("item > link")
```

```python
link_list[0].text
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
```

```python
news = []
for link in link_list:
    news_url = link.text
    news_reponse = requests.get(news_url)
    news_soup = BeautifulSoup(news_reponse.content, "html.parser")
    news_content = news_soup.select_one("#articlebody > .article_content")
    news_nouns = kkma.nouns(news_content.text)
    fitered_nouns = list(filter(lambda word: len(word)> 1 , kkma.nouns(news_content.text)))
    news.append(fitered_nouns)
```

```python
news
```

```python
from apyori import apriori
```

```python
rules = apriori(news, min_support=0.3, min_confidence=0.2)
results = list(rules)
```

```python
import pandas as pd
result_df = pd.DataFrame(None, columns=["lhs","rhs","support","confidence","lift"])
```

```python
index = 0
for row in results:
    support = row[1]
    ordered_stat = row[2]
    for ordered_item in ordered_stat:
        lhs = " ".join(x.strip() for x in ordered_item[0])
        rhs = " ".join(x.strip() for x in ordered_item[1])
        
        confidence = ordered_item[2]
        lift = ordered_item[3]
        result_df.loc[index] = [lhs,rhs,support,confidence,lift]
        index = index + 1
        
    
```

```python
result_df.sort_values(by="lift", ascending=False).head(5)
```

```python
temp = result_df.loc[(result_df.lhs.str.contains(""))]
```
