```python
import datetime
import cv2

#capture = cv2.VideoCapture(0)
capture = cv2.VideoCapture("../images/image_sample.mp4")
fcc = cv2.VideoWriter_fourcc('D','I','V','X')
record = False

width = int(capture.get(3))
height = int(capture.get(4))

while (capture.isOpened):
    ret, frame = capture.read()
    if ret == False:
        break
    cv2.imshow("VideoFrame", frame)

    now = datetime.datetime.now().strftime("%d_%H-%M-%S")
    key = cv2.waitKey(33)#1) & 0xFF
    

    if key == 27: # esc 
        break
    elif key == 26:   # ctrl + z
        print("")
        cv2.IMREAD_GRAYSCALE
        cv2.imwrite("./" + str(now) + ".png", frame)
        
        

    elif key == 24:  # ctrl + x
        print(" ")
        record = True
        outfps = 30
        #  
#         video = cv2.VideoWriter("./" + str(now) + ".avi", 
#                                 fcc, #  
#                                 outfps, 
#                                 (width, height))

        video = cv2.VideoWriter("./" + str(now) + ".avi", 
                                fcc, #  
                                outfps, 
                                (width, height),False)
    elif key == 3:
        print(" ") # ctrl + c
        record = False
        video.release()
    
    elif key == 1:
        print(" ") # ctrl + a
        cv2.IMREAD_GRAYSCALE
        frame = cv2.resize(src = frame, dsize  = (256,256))
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imwrite("./" + str(now) + ".png", gray)
    
    if record == True:
        print(" ..")
        video.write(frame)

capture.release()
cv2.destroyAllWindows()
```
